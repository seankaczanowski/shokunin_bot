#### Dataset Loading

```python
# Define a function for loading EURUSD_M15.csv dataset
def load_data_from_csv(csv_path):
    df = pd.read_csv(csv_path)
    df = df.rename(columns={
        "Open": "open", "High": "high", "Low": "low", "Close": "close"
    })
    df = df[['open', 'high', 'low', 'close']].astype(float)
    return df

# Load and preview your dataset
df = load_data_from_csv("datasets/EURUSD_M15.csv")
df.head()
```

#### Data Cleaning & Preparation

```python
# Calculate and append Ichimoku layers
from ta.trend import IchimokuIndicator

# Initialize IchimokuIndicator
ichimoku = IchimokuIndicator(
    high=df['high'],
    low=df['low'],
    window1=9,
    window2=26,
    window3=52,
    visual=True,
    fillna=False
)

# Add Ichimoku layers
df['tenkan_sen'] = ichimoku.ichimoku_conversion_line()
df['kijun_sen'] = ichimoku.ichimoku_base_line()
df['senkou_span_a'] = ichimoku.ichimoku_a()
df['senkou_span_b'] = ichimoku.ichimoku_b()
df['chikou_span'] = df['close'].shift(-26)
df.head()
```

```python
# Drop NaN and reset index
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# Define function for creating Heiken Ashi Candles
def compute_heiken_ashi(df):
    ha_df = pd.DataFrame(index=df.index)

    # Heiken Ashi Close is the average of open, high, low, close
    ha_df['ha_close'] = (df['open'] + df['high'] + df['low'] + df['close']) / 4

    # Heiken Ashi Open starts from the first candle
    ha_open = [(df['open'].iloc[0] + df['close'].iloc[0]) / 2]

    # Loop to compute remaining HA Open
    for i in range(1, len(df)):
        ha_open.append((ha_open[i-1] + ha_df['ha_close'].iloc[i-1]) / 2)

    ha_df['ha_open'] = ha_open

    # Heiken Ashi High = max(high, ha_open, ha_close)
    ha_df['ha_high'] = df[['high']].join(ha_df[['ha_open', 'ha_close']]).max(axis=1)

    # Heiken Ashi Low = min(low, ha_open, ha_close)
    ha_df['ha_low'] = df[['low']].join(ha_df[['ha_open', 'ha_close']]).min(axis=1)

    return ha_df
```

```python
# Calculate Heiken Ashi candles
ha_df = compute_heiken_ashi(df)

# Merge Heiken Ashi columns into the main dataframe
df = pd.concat([df, ha_df], axis=1)

df.head()
```

```python
# Define function for the life cycle logic of a trade
def simulate_trade_lifecycle(df, start_index, entry_price, direction, intent, ichimoku=None):
    trade = {
        "direction": direction,
        "entry_price": entry_price,
        "max_favorable_price": entry_price,
        "trail_armed": False,
        "trail_distance": 0.0020,
        "entry_mood": intent['mood'],
        "entry_confidence": intent['confidence'],
        "mood_strikes": 0,
        "kijun_strikes": 0,
        "doji_count": 0
    }

    for i in range(start_index + 1, min(start_index + 50, len(df))):
        price = df.iloc[i]['close']
        kijun = df.iloc[i]['kijun_sen']
        tenkan = df.iloc[i]['tenkan_sen']
        ha_open = df.iloc[i]['ha_open']
        ha_close = df.iloc[i]['ha_close']

        # Update max favorable price
        if direction == "bullish":
            trade['max_favorable_price'] = max(price, trade['max_favorable_price'])
        else:
            trade['max_favorable_price'] = min(price, trade['max_favorable_price'])

        moved = abs(trade['max_favorable_price'] - trade['entry_price'])
        if not trade['trail_armed'] and moved >= trade['trail_distance']:
            trade['trail_armed'] = True

        # Arm + Trigger trailing stop
        if trade['trail_armed']:
            if direction == "bullish" and price <= (trade['max_favorable_price'] - trade['trail_distance']):
                return i, price, "Trailing Stop", round((price - trade['entry_price']) * 10000, 1)
            if direction == "bearish" and price >= (trade['max_favorable_price'] + trade['trail_distance']):
                return i, price, "Trailing Stop", round((trade['entry_price'] - price) * 10000, 1)

        # Exit if price lingers on wrong side of Kijun
        if direction == "bullish" and price < kijun:
            trade['kijun_strikes'] += 1
        elif direction == "bearish" and price > kijun:
            trade['kijun_strikes'] += 1
        else:
            trade['kijun_strikes'] = 0

        if trade['kijun_strikes'] >= 3:
            pnl = price - trade['entry_price'] if direction == "bullish" else trade['entry_price'] - price
            return i, price, "Kijun Drift Exit", round(pnl * 10000, 1)

        # Optional: Mood shift conflict exit
        try:
            candle_dicts = [
                {"open": row.open, "high": row.high, "low": row.low, "close": row.close}
                for row in df.iloc[:i+1].itertuples()
            ]
            cloud_top = df.iloc[i]['senkou_span_a']
            cloud_bottom = df.iloc[i]['senkou_span_b']
            intent_now = get_intent(candle_dicts, cloud_top, cloud_bottom)

            if mood_conflict(trade['entry_mood'], trade['entry_confidence'], intent_now['mood'], trade['direction']):
                trade['mood_strikes'] += 1
                if trade['mood_strikes'] >= 2:
                    pnl = price - trade['entry_price'] if direction == "bullish" else trade['entry_price'] - price
                    return i, price, "Mood Conflict Exit", round(pnl * 10000, 1)
            else:
                trade['mood_strikes'] = 0
        except:
            pass

        # Exit if 2+ small-body (doji-like) HA candles in a row
        ha_body = abs(ha_close - ha_open)
        if ha_body < 0.0003:
            trade['doji_count'] += 1
        else:
            trade['doji_count'] = 0

        if trade['doji_count'] >= 2:
            pnl = price - trade['entry_price'] if direction == "bullish" else trade['entry_price'] - price
            return i, price, "HA Indecision Exit", round(pnl * 10000, 1)

    # Final fallback (last resort)
    final_price = df.iloc[min(start_index + 50, len(df) - 1)]['close']
    pnl = final_price - trade['entry_price'] if direction == "bullish" else trade['entry_price'] - final_price
    return min(start_index + 50, len(df) - 1), final_price, "Final Timeout Exit", round(pnl * 10000, 1)
```

```python
# Define a function for computing Heiken Ashi trend strength
def ha_trend_strength(df, recent_n=5):
    """
    Compute Heiken Ashi trend strength from precomputed HA candles.
    Assumes df has 'ha_open' and 'ha_close' columns.
    Returns:
        1  → Bullish trend
        -1 → Bearish trend
         0 → Neutral or weak
    """
    recent = df.iloc[-recent_n:]

    bullish = (recent['ha_close'] > recent['ha_open']).sum()
    bearish = (recent['ha_close'] < recent['ha_open']).sum()

    if bullish == recent_n:
        return 1
    elif bearish == recent_n:
        return -1
    else:
        return 0
```

```python
# Define a function for the simulation of a trading strategy
def run_simulation_from_df(df, instrument="EUR_USD", lookback_window=100, show_logs=True):
    trades = []
    i = lookback_window + 52  # Ensure Ichimoku indicators are valid

    while i < len(df):
        # Slice recent candles for intent engine
        candles_recent = [
            {"open": row.open, "high": row.high, "low": row.low, "close": row.close}
            for row in df.iloc[i - lookback_window : i].itertuples()
        ]

        try:
            cloud_top = df.iloc[i]['senkou_span_a']
            cloud_bottom = df.iloc[i]['senkou_span_b']
            intent = get_intent(candles_recent, cloud_top, cloud_bottom)
        except Exception as e:
            if show_logs:
                print(f"[{i}] get_intent failed: {e}")
            i += 1
            continue

        # Use precomputed HA candles
        ha_slice = df.iloc[i - lookback_window : i]
        ha_trend = ha_trend_strength(ha_slice)

        if show_logs:
            print(f"[{i}] Intent: {intent['type']} | Should trade: {intent['should_trade']} | HA Trend: {ha_trend}")

        if intent["should_trade"]:
            ha_confirms = (
                (intent["type"] == "bullish_bias" and ha_trend == 1) or
                (intent["type"] == "bearish_bias" and ha_trend == -1)
            )

            if ha_confirms:
                entry_price = candles_recent[-1]["close"]
                direction = "bullish" if intent["type"] == "bullish_bias" else "bearish"

                ichimoku = {
                    "tenkan_sen": list(df['tenkan_sen']),
                    "kijun_sen": list(df['kijun_sen']),
                    "senkou_span_a": list(enumerate(df['senkou_span_a'])),
                    "senkou_span_b": list(enumerate(df['senkou_span_b'])),
                    "chikou_span": list(df['chikou_span']),
                }

                exit_idx, exit_price, reason, pnl = simulate_trade_lifecycle(
                    df, i, entry_price, direction, intent, ichimoku
                )

                trades.append({
                    "entry_index": i,
                    "exit_index": exit_idx,
                    "instrument": instrument,
                    "direction": direction,
                    "entry_price": entry_price,
                    "exit_price": exit_price,
                    "exit_reason": reason,
                    "pnl_pips": pnl,
                    "confidence": intent["confidence"],
                    "mood": intent["mood"],
                    "ha_trend": ha_trend
                })

                if show_logs:
                    print(f"{i} → {exit_idx} | {direction.upper()} | {entry_price:.5f} → {exit_price:.5f} | {pnl:+.1f} pips | {reason}")

                i = exit_idx + 1
                continue

        i += 1

    return pd.DataFrame(trades)
```
